#### Objetivo
Explorar el comportamiento del material particulado, en diferentes situaciones 'controladas'.

#### Metodología
1. (*0h 0min*)  -> Encender las dos estaciones, la una al lado de la otra, con la habitación cerrada.
2. (*0h 15min*) -> Encender un recuadro de papel de (5x5 cm), esparcir lo más uniformemente posible en la habitación (cerrada).
3. (*1h 0min*)  -> Abrir la puerta y con un ventilador expulsar el material particulado.
4. (*2h 0min*)  -> Apagar el ventilador y cerrar la puerta.
5. (*2h 0min*)  -> **Ubicar la estación *outdoor* fuera de la habitación.**
6. (*2h 15min*) -> Encender un recuadro de papel de (5x5 cm), esparcir lo más uniformemente posible en la habitación (cerrada).
    * 6.1 (*2h 17min*) -> Abrir la puerta.
7. (*3h 15min*) -> Con el ventilador expulsar el material particulado.
8. (*4h 15min*) -> Encender un recuadro de papel de (5x5 cm), esparcir lo más uniformemente posible en la habitación (cerrada).
    * 8.1 (*4h 17min*) -> Abrir la puerta y la ventana.
9. (*5h 15min*) -> Con el ventilador expulsar el material particulado.
10. (*6h 15min*) -> Encender un recuadro de papel de (5x5 cm), esparcir lo más uniformemente posible en la habitación (cerrada).
    * 10.1 (*6h 17min*) -> Abrir la puertaa y encender el ventilador
11. Fin de la prueba
